```python
import pandas as pd
```

```python
orders = pd.read_csv('orders.csv', sep=';')
products = pd.read_csv('products.csv', sep=';')
```

```python
orders.head()
```

```python
products.head()
```

```python
orders_product = orders.merge(products, left_on='ID товара', right_on='Product_ID', how='left')
```

```python
orders_product.head()
```

```python
orders_product.tail()
```

```python
orders_product[orders_product['Статус'] == 'Отменён']['Name']
```

```python
mask = orders_product['Оплачен'] == 'Да'
orders_product['profit'] = orders_product['Количество'] * orders_product['Price']
orders_product[mask].groupby('ID Покупателя')['profit'].sum().sort_values(ascending=False)
```
